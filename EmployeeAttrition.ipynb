Importing the Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Loading the Datasets

```python
data=pd.read_csv('Attrition.csv')
data.head()
```

Setting the Employee ID as the index since it is unique

```python
data=data.set_index('EmployeeID')
data.head()
```

Viewing the dataset and its specifications

```python
data.describe()
```

```python
data.info()
```

Checking to see if the dataset has any null values

```python
data.isnull().sum()
```

Filling the Null values in the datset with the mean and the median values 

```python
data['NumCompaniesWorked']=data['NumCompaniesWorked'].fillna(data['NumCompaniesWorked'].mean())
data['TotalWorkingYears']=data['TotalWorkingYears'].fillna(data['TotalWorkingYears'].median())
data
```

```python
data.isnull().sum()
```

Now we will fill the features such as Environment Satisfaction, Job Satisfaction and Work-Life Balance with their mode values to obtain a realistic represenation of the realistic scenario

```python
data['EnvironmentSatisfaction']=data['EnvironmentSatisfaction'].fillna(data['EnvironmentSatisfaction'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data['JobSatisfaction']=data['JobSatisfaction'].fillna(data['JobSatisfaction'].mode()[0])
data['WorkLifeBalance']=data['WorkLifeBalance'].fillna(data['WorkLifeBalance'].mode()[0])
data
```

```python
data.isnull().sum()
```

We have now ensured that our datset no longer has any null values

```python
data['Over18'].value_counts()
```

As we can see the above feature is the same for every single employee, so it goes without saying that all features similar to it such as Over18, StandarHours, EmployeeCount can be dropped. Similarly the TotalWorkingYears field has extremely low correlation with the Attrition Statistics

```python
data.drop(columns='Over18',inplace=True)
```

```python
data.drop(columns='StandardHours',inplace=True)
```

```python
data.drop(columns='TotalWorkingYears',inplace=True)
```

```python
data.drop(columns='EmployeeCount',inplace=True)
```

Let's see the correlation between the most important parameters and the attrition field.

```python
new=data.drop(columns=['Age','Department','Education','EducationField','Gender','JobRole','MaritalStatus','NumCompaniesWorked'])
new.head()
```

```python
new.drop(columns='BusinessTravel',inplace=True)
new.head()
```

Label Encoding of the Attrition Column values since a heatmap takes only numerical inputs.

```python
new.loc[new.Attrition=='Yes','Attrition']=1
new.loc[new.Attrition=='No','Attrition']=0
```

```python
plt.figure(figsize=(11,6))
sns.heatmap(new.corr(),annot=True,linewidth=1)
plt.title('Correlation Between Parameters',fontsize=20)
plt.show()
```

Attrition on the basis of Job Satisfaction, Environment Satisfaction and Work-Life Balance

```python
# Set the style for Seaborn plots
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# Plot for Job Satisfaction
sns.countplot(x='JobSatisfaction', hue='Attrition', data=data, ax=axes[0])
axes[0].set_title('Job Satisfaction vs Attrition')

# Plot for Environment Satisfaction
sns.countplot(x='EnvironmentSatisfaction', hue='Attrition', data=data, ax=axes[1])
axes[1].set_title('Environment Satisfaction vs Attrition')

# Plot for Work-Life Balance
sns.countplot(x='WorkLifeBalance', hue='Attrition', data=data, ax=axes[2])
axes[2].set_title('Work-Life Balance vs Attrition')

# Adjust layout
plt.tight_layout()
plt.show()
```
